# Teorema de Bayes:

Fórmula:

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} $$

Onde:
- **P(A|B)** é a probabilidade de A ocorrer dado que B ocorreu.
- **P(B|A)** é a probabilidade de B ocorrer dado que A ocorreu.
- **P(A)** e **P(B)** são as probabilidades de A e B ocorrerem independentemente.

Agora, vamos a um exemplo do dia a dia para entender melhor:

Imagine que você tem uma bolsa com 10 bolas: 7 são azuis e 3 são vermelhas. Agora, você coloca a mão na bolsa e pega uma bola sem olhar. Qual é a probabilidade de você pegar uma bola azul? Isso é o que chamamos de **P(A)**, a probabilidade de A (pegar uma bola azul) ocorrer independentemente. Nesse caso, **P(A)** é 7/10 ou 0,7.

Agora, digamos que alguém te diz que a bola que você pegou não é vermelha. Isso é o que chamamos de evento B. Agora, você quer saber qual é a probabilidade de a bola ser azul, dado que você sabe que ela não é vermelha. Isso é o que chamamos de **P(A|B)**.

Para calcular isso, precisamos saber **P(B|A)**, a probabilidade de a bola não ser vermelha, dado que é azul. Como todas as bolas azuis não são vermelhas, **P(B|A)** é 1.

Finalmente, precisamos saber **P(B)**, a probabilidade de a bola não ser vermelha. Como temos 7 bolas azuis e 3 bolas vermelhas, **P(B)** é 7/10 ou 0,7.

Agora podemos calcular **P(A|B)** usando a fórmula do Teorema de Bayes:

$$ P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)} = \frac{1 \cdot 0,7}{0,7} = 1 $$

Portanto, se sabemos que a bola não é vermelha, a probabilidade de ela ser azul é 1, ou 100%.
